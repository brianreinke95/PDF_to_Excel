## Separar horas de la descripción.

```python
import re as regex


def separador(string):
    hours = regex.findall(r'\d\.\d{2}', string)
    if hours:
            hours = hours.pop(0)
            string = string.replace(hours, '')
    else:
        hours = np.NaN
    
    string = string.lstrip()
    return string, hours
```

```python
df[9].dropna(subset=[2], inplace=True)
aux = df[9]
aux = aux.apply(lambda x: separador(x[2]), axis=1)
aux = pd.Series(aux).apply(pd.Series)
aux
```
