## Date y fecha. primer caso

```python
# Si comienza con una fecha hacer split con 2. Sino con 1
# Al tener las listas se dividirán en 3 columnas. Los que tengan solo 2 se llenará con NaN la de la fecha y luego
# se hará un ffill.
import re as regex


def separador(string):
    
    date = regex.findall(r'^\d{1,2}/\d{1,2}/\d{4}', string)
    
    tag = regex.findall(r'\b[A-Z]{2,3}', string)
    
    if date:
            date = date.pop(0)
            string = string.replace(date, '')
    else:
        date = np.NaN
    if tag:
            tag = tag.pop(0)
            string = string.replace(tag, '')
    else:
        tag = np.NaN
    
    string = string.lstrip()
    return date, tag

```

```python
aux = df[0].dropna(subset=['Unnamed: 0'])
aux = aux.apply(lambda x: separador(x['Unnamed: 0']), axis=1)
aux = aux.apply(pd.Series)
aux.columns = ['date', 'attorney']
aux.dropna(how="all", inplace=True)
aux.fillna(method='ffill', inplace=True)
aux
# separador(df[0].loc[0, 'Unnamed: 0'])
```

## Dejando igual al df 0 y 7 con los otros

```python
import re as regex
def separador_text(string):
    
    date = regex.findall(r'^\d{1,2}/\d{1,2}/\d{4}', string)
    
    tag = regex.findall(r'\b[A-Z]{2,3}', string)
    if date:
            date = date.pop(0)
            string = string.replace(date, '')
    else:
        date = np.NaN
    if tag:
            tag = tag.pop(0)
            string = string.replace(tag, '')
    else:
        tag = np.NaN
    
    string = string.lstrip()
    return string, tag
```

```python
df[0] = df[0].dropna(subset=['Unnamed: 0']).reset_index(drop=True)
df[7] = df[7].dropna(subset=['Unnamed: 0']).reset_index(drop=True)
```

```python
aux = df[0]
itera = map(separador_text, aux.loc[:, 'Unnamed: 0'])
df[0][['Unnamed: 1', 'Unnamed: 0']] = pd.Series(itera).apply(pd.Series)
itera = map(separador_text, df[7].loc[:, 'Unnamed: 0'])
df[7][['Unnamed: 1', 'Unnamed: 0']] = pd.Series(itera).apply(pd.Series)
```
