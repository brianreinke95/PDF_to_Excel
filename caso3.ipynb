# Date

```python
Date = aux['Date'].dropna()
```

```python
import re as regex
def remove_no_dates(string):
    if regex.search(r'\d{1,2}\.\d{1,2}\.\d{1,2}', string):
        return string
    else:
        return np.NaN

asd = [remove_no_dates(x) for x in Date]
Date = pd.Series(asd).dropna()
Date = pd.to_datetime(Date, format='%y.%m.%d').dt.strftime('%m/%d/%Y')
```

```python
for date in Date:
    print(date)
```
