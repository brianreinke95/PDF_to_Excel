```python
import re as regex
def separador(string):
    
    date = regex.findall(r'^\d{1,2}/\d{1,2}/\d{4}', string)
    
    tag = regex.findall(r'\b[A-Z]{2,3}', string)
    
    if date:
            date = date.pop(0)
            string = string.replace(date, '')
    else:
        date = np.NaN
    if tag:
            tag = tag.pop(0)
            string = string.replace(tag, '')
    else:
        tag = np.NaN
    
    string = string.lstrip()
    return date, tag, string
```
